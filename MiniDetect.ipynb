```
# MiniDetect est une demo qui montre qu'une appli de détection de forme 
# peut être définie en 15 lignes. Mais évidemment en dessous de ces 15 lignes 
#...il y en a beaucoup d'autres

#@title
!pip install opencv-python
!pip install cvlib
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox
%matplotlib inline
import matplotlib.pyplot as plt

def showim(img):

  im = cv2.imread(img)
  im= cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
  bbox, label, conf = cv.detect_common_objects(im) 
  output= draw_bbox(im, bbox, label, conf) 
  plt.imshow(output)
  plt.show()
```
